# Group by Function

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby(['Team']).groups #grouping the indexs using team column
```

```python
#we can get single group using its elements.
#the function used is get_group
```

```python
grouped = df.groupby('Team')
grouped.get_group('Kings') #shwing just the elemwnts of the group called kings
```

```python
df.groupby('Team').get_group('Kings')
```

# How can we use group by function
1. For aggregation: We can apply aggregation on the groups after using group by function

.agg is the function
    

```python
df
```

```python
#suppose we want mean of the points which belongs to year 2017 only


# we will group by using year then we will apply mean on points of the year 2014
```

```python
df.groupby('Year').get_group(2017)['Points'].agg(np.mean)
```

```python
df.groupby('Year').get_group(2017)['Points'].agg([np.mean,np.std,np.sum]) #we can also get multiple aggreagation 
```

2. For Transforming the data.

.transform is the function

```python
#the data is first grouped and then transformed  as per the needs

#Suppose we want to transform the values of column- Points into value minus mean of value where the rank is 1.:
here we will first group the data and then transform it
```

```python
#creating a tranforming def
trans=lambda x:x-x.mean()
```

```python
df.groupby('Rank').get_group(1)['Points'].transform(trans)
```

3. Filteration: We can group by and then use filteration method to filter the data as per our requirments

```python
#suppose we want points where rank is 4 times 
```

```python
df.groupby('Rank')['Points'].filter(lambda x: len(x)==4) 
```

# Merge in the Pandas

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
left
```

```python
right
```

# ON parameter: It picks the common attributes only . On is used to select the exact column name as the key.

```python
pd.merge(left,right,on=['id','subject_id'])
```

# HOW parameter : It picks the dataframe as the key.

It has left, right , inner, outer join

```python
pd.merge(left,right,on='subject_id',how='left') 
#it will take all the values of left dataframe and common values of the right also
```

```python
pd.merge(left,right,on='subject_id',how='right') 
```

```python
pd.merge(left,right,on='subject_id',how='outer') 
```

```python
pd.merge(left,right,on='subject_id',how='inner')  #same as just using on as it picks the common value of the key only.
```

```python
pd.merge(left,right,on='subject_id') #same as above
```

# Concatenate in Pandas: 

It again involves joining series or df but as a whole. It does not take into account common attributes or anything of that sort. It just means putting two df together be it row wise or column wise


```python
pd.concat([left,right]) #default is axis=0 that is column wise
```

```python
pd.concat([left,right],axis=1)
```

```python
pd.concat([left,right],axis=0,ignore_index=True) #ignoring index means you want to give fresh index
```

```python
# Append also works like concate only but concat in axis=0
left.append(right)
```
